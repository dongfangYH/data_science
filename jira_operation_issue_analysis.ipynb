```python
import numpy as np
import pandas as pd
import glob
import os
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 328)
```

```python
cols_with_parent = ['Summary', 'Issue key', 'Issue Type', 'Status', 'Priority', 'Resolution', 'Assignee', 'Reporter', 
            'Creator', 'Created', 'Resolved', 'Description', 'Environment','Custom field (Requested From:)','Custom field ([CHART] Date of First Response)',
            'Parent', 'Status Category', 'Status Category Changed']
cols_without_parent = cols_with_parent.copy()
cols_without_parent.remove('Parent')
```

```python
file_prefix='JIRA_operation_issue'
file_pattern = os.path.join("./data/jira2/", f'{file_prefix}*.csv')
csv_files = glob.glob(file_pattern)

dataframes = []

for file in csv_files:
    df_sample = pd.read_csv(file, nrows=0)
    if 'Parent' in df_sample.columns:
        column_to_read = cols_with_parent
    else:
        column_to_read = cols_without_parent
    dataframes.append(pd.read_csv(file, usecols=column_to_read))
df = pd.concat(dataframes, ignore_index=True)
df.head()
```

```python
df.info()
```

```python
df.drop(columns=['Parent', 'Environment'], inplace=True)
df =df.rename(columns={
    'Custom field ([CHART] Date of First Response)': 'First Response'
})
df['Created'] = pd.to_datetime(df['Created'])
df['Resolved'] = pd.to_datetime(df['Resolved'])
df['First Response'] = pd.to_datetime(df['First Response'])
```
